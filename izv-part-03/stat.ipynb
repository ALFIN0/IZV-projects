## Hypotéza 1:
**Na cestách 1. triedy sa pri nehodách zomieralo rovnakou pravdepodobnosťou ako na cestách 3. triedy.**

Import knižníc

```python
import pandas as pd
from scipy.stats import ttest_ind, chi2_contingency
```

Načítanie dataframe

```python
df = pd.read_pickle("accidents.pkl.gz")
df
```

Ktoré nehody boli smrteľné na cestach prvej triedy (1) a tretej triedy (3)

```python
df_class_1_3 = df[(df.p36 == 1) | (df.p36 == 3)]

df_class_1_3["tragic"] = df_class_1_3.apply(lambda x: x.p13a > 0,axis=1)

df_class_1_3
```

Konfiguracna tabuľka

```python
contigency = pd.crosstab(df_class_1_3["tragic"], df_class_1_3["p36"]) 
contigency
```

Výpočet pravdepodobnosti podľa chi $\chi^2$ 




```python
c, p, dof, expected = chi2_contingency(contigency)
p
```

3,54 e-29 je menšie ako 0,05 takže zamietame hypotézu 1 => na cestách 1. triedy sa zomiera viac ako na cestách 3. triedy
